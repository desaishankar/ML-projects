```python
model_mean_pred = df.priceModLog.mean()
```

```python
df.plot(kind="line", x="date", y = ["priceMod", "priceMean"])
```

```python
def RMSE(predicted, actual):
    mse = (predicted - actual)**2
    rmse = np.sqrt(mse.sum()/mse.count())
    return rmse
```

```python

model_mean_RMSE = RMSE(df.priceMean, df.priceMod)
model_mean_RMSE
```

```python
#split between the training and the test data sets. 
df_train = dfmumbai.iloc[0:117]
df_test = dfmumbai.iloc[117:]
```

```python
df_train
```

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
y_hat_avg = df_test.copy()
fit1 = ExponentialSmoothing(np.asarray(df_train['priceMod']) ,seasonal_periods=7 ,trend='add', seasonal='add',).fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(len(df_test))
plt.figure(figsize=(15,15))
plt.plot( df_train['priceMod'], label='Train')
plt.plot(df_test['priceMod'], label='Test')
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(df_test.priceMod, y_hat_avg.Holt_Winter))
print(rms)
```
